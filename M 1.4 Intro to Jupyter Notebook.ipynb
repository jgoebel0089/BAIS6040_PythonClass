
## Variables and Data Types

You can store values in **variables** and use them in calculations.
Python automatically assigns a type (int, float, string, etc.) based on the value.


```python

name = "Alex"
age = 25
height = 1.78

print("Name:", name)
print("Age:", age)
print("Height:", height, "meters")
print("Type of age variable:", type(age))
```


## Basic Math Operations

Python supports basic arithmetic operations:
- Addition (`+`)
- Subtraction (`-`)
- Multiplication (`*`)
- Division (`/`)
- Exponentiation (`**`)


```python

x = 10
y = 3

add = x + y
div = x / y
exp = x ** y

print("Addition:", add)
print("Division:", div)
print("Exponentiation:", exp)
```
