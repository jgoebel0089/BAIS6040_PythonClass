# Notebook Demonstration 1.9 Debugging with Python
This notebook demonstrates 5 basic Python errors and how to fix them.

```python

# 1. Create a list of numbers from 1 to 10
numbers = list(range(1, 11))

# 2. Print the original list
print("Original list:", numbers)

# 3. Append a new number
numbers.append(11)

# 4. Insert a number at index 5
numbers.insert(5, 99)

# 5. Remove the number 99
numbers.remove(99)

# 6. Pop the last element
last = numbers.pop()
print("Popped:", last)

# 7. Reverse the list
numbers.reverse()

# 8. Sort the list
numbers.sort()

# 9. Slice the first 5 elements
first_five = numbers[:5]
print("First five:", first_five)

# 10. Check if 5 is in the list
print("Is 5 in list?", 5 in numbers)

# 11. Count occurrences of 5
print("Count of 5:", numbers.count(5))

# 12. Create a list of squares
squares = [x**2 for x in numbers]

# 13. Print squares
print("Squares:", squares)

# 14. Filter even numbers
evens = [x for x in numbers if x % 2 == 0]

# 15. Print evens
print("Even numbers:", evens)

# 16. Define a function to double a number
def double(x):
    return x * 2

# 17. Apply function to list
doubled = [double(x) for x in numbers]

# 18. Print doubled list
print("Doubled:", doubled)

# 19. Create a nested list
nested = [[x, x**2] for x in numbers]

# 20. Print nested list
print("Nested list:", nested)

# 21. Flatten nested list
flat = [item for sublist in nested for item in sublist]

# 22. Print flattened list
print("Flattened:", flat)

# 23. Zip two lists
zipped = list(zip(numbers, squares))

# 24. Print zipped list
print("Zipped:", zipped)

# 25. Unzip the list
nums, sqs = zip(*zipped)

# 26. Print unzipped lists
print("Nums:", nums)
print("Sqs:", sqs)

# 27. Create a list of strings
words = ["apple", "banana", "cherry"]

# 28. Convert to uppercase
upper_words = [word.upper() for word in words]

# 29. Print uppercase words
print("Uppercase:", upper_words)

# 30. Join words into a sentence
sentence = " ".join(words)

# 31. Print sentence
print("Sentence:", sentence)

# 32. Split sentence back to list
split_words = sentence.split()

# 33. Print split words
print("Split words:", split_words)

# 34. Enumerate words
for i, word in enumerate(words):
    print(f"{i}: {word}")

# 35. Check length of list
print("Length of numbers:", len(numbers))

# 36. Clear the list
numbers.clear()

# 37. Print cleared list
print("Cleared list:", numbers)

# 38. Refill list
numbers = list(range(1, 6))

# 39. Multiply list
repeated = numbers * 2

# 40. Print repeated list
print("Repeated list:", repeated)
```

### 1. SyntaxError
Occurs when Python code is not written correctly.

```python
# Missing colon in if statement
if True
    print('Hello')

# Corrected version:
# if True:
#     print('Hello')
```

### 2. NameError
Occurs when a variable is used before it is defined.

```python
print(x)  # x is not defined

# Corrected version:
# x = 10
# print(x)
```

### 3. TypeError
Occurs when an operation is applied to an object of inappropriate type.

```python
result = 'Age: ' + 25  # Cannot concatenate str and int

# Corrected version:
# result = 'Age: ' + str(25)
# print(result)
```

### 4. IndexError
Occurs when trying to access an index that is out of range.

```python
numbers = [1, 2, 3]
print(numbers[5])  # Index out of range

# Corrected version:
# numbers = [1, 2, 3]
# print(numbers[2])
```

### 5. ValueError
Occurs when a function receives an argument of the right type but inappropriate value.

```python
int('abc')  # Cannot convert string to int

# Corrected version:
# int_value = int('123')
# print(int_value)
```
