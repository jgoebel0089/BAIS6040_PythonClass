```python
returns = [23.31,24.23,-19.44,26.89,16.26,28.88,-6.24,19.42,9.54,-0.73,11.39,29.60,13.41,0.00,12.78,23.45,-38.49,3.53,13.62,3.00,8.99,26.38,-23.37,-13.04,-10.14,19.53,26.67]
sp = sum(returns) / len(returns) / 100

limits = [2000,2000,2000,2000,3000,3000,3000,4000,4000,4000,5000,5000,5000,5000,5500,5500,5500,5500,5500,5500,6000,6000,6000,6000,6500,7000,7000]
avg_increase = (max(limits) - min(limits)) / len(limits)
interval = round(500 / avg_increase)

def roth_simulator(yearly_return, interval, contribution=7000, years=35):
    catch_up = 0
    balance = 0
    for x in range(years):
        if (x+2) % interval == 0:
            contribution += 500
        if x > 24:
            catch_up = 1000
        balance += contribution
        balance += catch_up
        balance = balance * (1 + yearly_return)
    return balance

bal = roth_simulator(sp, interval)
print("sp =", sp)
print("interval =", interval)
print("Final balance = $", round(bal, 2))
```
