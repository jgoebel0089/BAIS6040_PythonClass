1\. (5 pt) Create a CSV file named `hw_diamonds.csv` in the `outcome` directory. The file contains 7 column names `carat`, `cut`, `color`, `clarity`, `depth`, `table`, and `price` in the first row and the values from the corresponding columns of `df` in the following rows. The delimiter is a slash (/) for columns and a new line (\n) for rows. For example, the first 3 rows in the file must look as below. Do not use pandas.to_csv. Hint - loop through each row of the dataframe above using df.iterrows(). Use format printing .format or f-strings. https://www.geeksforgeeks.org/python/python-output-formatting/

```python
# Your answer here
columns = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price']
with open("outcome/hw_diamonds.csv", 'w') as f:
    # Write header
    f.write('/'.join(columns) + '\n')
    
    # Loop through each row in the DataFrame
    for index, row in df.iterrows():
        # Create a slash-separated string of values from the specified columns
        line = '/'.join(str(row[col]) for col in columns)
        f.write(f"{line}\n")

```

```python
# Check your answer here. (Do not make any change to this cell. Just run this cell.)
os.path.isfile("outcome/hw_diamonds.csv")                # Expected output: True
```

2\. (5 pt) Read the file `hw_diamonds.csv` created in Question 1 and print the first 10 rows in the file including the header row, separated by a comma (,) between columns. For example, the output must be printed as below. Do not use Pandas to read in the file.

```python
# Your answer here
with open("outcome/hw_diamonds.csv", 'r') as fh:
    # Read all lines
    lines = fh.readlines()

    # Print the first 10 rows
    for line in lines[:10]:
        print(line.strip().replace('/', ','))
```

3\. (5 pt) Create a list `diamonds`, in which each element is a dictionary with 3 keys being `carat`, `cut`, and `color` and the values being their corresponding values in `df`. For example, the first element in `diamonds` must look as below. Then, serialize the first 1,000 items in `diamonds` as a JSON formatted stream to a file named `hw_diamonds.json` saved in the `outcome` directory.

```python
# Your answer here
import os
import json

diamonds = []

for idx, row in df.iterrows():
    diamond = {
        'carat': row['carat'],
        'cut': row['cut'],
        'color': row['color']
    }
    diamonds.append(diamond)

with open("outcome/hw_diamonds.json", "w") as fh:
    json.dump(diamonds[:1000], fh, indent=2)
```
