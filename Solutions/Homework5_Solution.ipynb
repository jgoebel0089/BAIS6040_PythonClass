1\. (3 pts) Plot a box plot that displays the distributions of the 3 columns `x`, `y`, and `z` in `dfdi`. Set the `figsize` to 15 x 7. Customize other arguments, including title, labels, alignment and so on. 

```python
# Your answer here

import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(15, 7))


data = [dfdi['x'], dfdi['y'], dfdi['z']]
labels = ['x', 'y', 'z']

# Create the box plot using matplotlib
plt.figure(figsize=(15, 7))
plt.boxplot(data, labels=labels, patch_artist=True)


# Customize the plot
plt.title('Distribution of Dimensions x, y, and z in Diamonds Dataset', fontsize=16)
plt.xlabel('Dimension', fontsize=14)
plt.ylabel('Value (mm)', fontsize=14)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

2\. (4 pts) Plot a histogram that plots the distribution of `price` in `dfdi` with 10 bins. Customize the figuresize and other arguments, including title, labels, alignment and so on.

```python
# Your answer here

plt.figure(figsize=(15, 7))

plt.hist(dfdi['price'], bins=10, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Distribution of Diamond Prices', fontsize=16)
plt.xlabel('Price (USD)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

3\. (5 pts) There are 7 unique colors in the `color` column of `dfdi`: *D*, *E*, *F*, *G*, *H*, *I*, and *J*. Plot a pie chart and a bar plot in one figure, two columns in one row. The pie chart displays the distribution(percentage) of the count of each color and the bar plot displays the mean prices for the 7 colors. Set the `figsize` to 14 x 7. Use the <b>groupby</b> method to calculate the mean prices for the 7 colors. 

```python
# Your answer here

color_counts = dfdi['color'].value_counts().sort_index()
mean_prices = dfdi.groupby('color')['price'].mean().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].pie(color_counts, labels=color_counts.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Distribution of Diamond Colors')

axes[1].bar(mean_prices.index, mean_prices.values, color='skyblue', edgecolor='black')
axes[1].set_title('Mean Price by Diamond Color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Mean Price (USD)')
axes[1].grid(True)

plt.show()
```

5\. Plot a Pie Chart to show the total number of passergers survived or not (died). Set the `startangle` as 90, add the percentage label inside the pie with the format `%.1f%%`, and split the two parts by setting the `explode` as \[0, 0.1\], set the color as "green" for survived group and "gray" for non-survived group. 

```python
# Your answer here
# Fill your code below
count = [dft.survived[dft.survived==1].sum(), dft.survived[dft.survived==0].shape[0]]

plt.pie(count, labels=['Survived', 'Died'], startangle=90,
       autopct="%.1f%%", explode=[0, 0.1], colors=["green", "gray"] )

# Show the plot
plt.show()
```

6\. Plot two pie charts, two columns in one row. Set the `figsize` to 14 x 7. The left pie chart shows the passenger class `Pclass` distribution for the survived passengers; the pie chart on the right shows the `Pclass` distribution for passengers who didn't survive. Add the text labels (given in the `labels` list) and the percentage label inside each pie chart with the format `%.1f%%`, and add a title for each pie chart.

```python
pclass_s = dft[dft.survived==1].pclass.value_counts().sort_index()
```

```python
pclass_s
```

```python
# Your answer here
# Fill your code below
fig, axs = plt.subplots(1, 2, figsize=(14,7))

ax1 = axs[0]
ax2 = axs[1]

pclass_s = [dft[(dft.survived==1) & (dft.pclass==1)].shape[0], dft[(dft.survived==1) & (dft.pclass==2)].shape[0], dft[(dft.survived==1) & (dft.pclass==3)].shape[0]]
pclass_d = [dft[(dft.survived==0) & (dft.pclass==1)].shape[0], dft[(dft.survived==0) & (dft.pclass==2)].shape[0], dft[(dft.survived==0) & (dft.pclass==3)].shape[0]]

labels = ['class 1', 'class 2', 'class 3']

# Pie chart on the left is for passengers who survived
ax1.pie(pclass_s, labels=labels, autopct="%.1f%%")
ax1.set_title("Survived")

# Pie chart on the right for passengers who died
ax2.pie(pclass_d, labels=labels, autopct="%.1f%%")
ax2.set_title("Died")

# Show the plot
plt.show()
```
