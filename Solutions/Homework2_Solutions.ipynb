Set up if/elif/else statements that returns the balance and warning strings if `contribution` is > 583.33 or < 0. Otherwise, it returns the money you have left, `balance` - `contribution` and the f-string of the future value of your investment for 1 month 30 years from now! Assume a 10\% annual rate of return and use the future value formula.

https://www.w3schools.com/python/python_string_formatting.asp

https://www.investopedia.com/terms/f/futurevalue.asp

```python
#In this cell define your roth_ira function
def roth_ira (balance):
    contribution = float(input("How much will you contribute to your Roth IRA? $"))
    fv = contribution * (1+.1)**30
    if contribution < 0:
        return balance, "You can't contribute a negative value."
    elif contribution > 583.33:
        return balance, "You can't contribute more than $583.33"
    else:
        return balance - contribution, f"You will have ${fv:.2f} in 30 years if you invest ${contribution:.2f} now."
```

5\. (8 pts) You will be creating a loan_calculator function that calculates the student loan monthly payment shown at this website. We will assume a standard repayment plan of 10 years and 120 payments.

https://lendedu.com/blog/student-loan-payment-calculator/

Outside of the function, get the annual rate (.07 for 7 percent) and principal.
When defining the function, you have 3 arguments. The first argument is your `balance`, which is the same as Problem 3 or what is left in your monthly budget. The other 2 arguments are `rate` and `loan_amt`, which we got as input from the user in the cell before. The default arguments for those will be .075 and 20000.

Inside of the function, calculate the monthly loan payment based on the calculation from the website above. Return `monthly_payment` and `money_left` on your budget after you pay your student loan monthly payment.

```python
annual_rate = float(input("Please input your interest rate as a decimal (ex. .1 for 10 percent): "))
principal = float(input("How much did you originally take out for your loan? "))
```

```python
def loan_calculator(balance, rate=0.075, loan_amt=20000):
    interest = (1 + rate/365.25)**(365.25*10)
    monthly_payment = loan_amt*(rate/365.25*interest)/(interest-1)*30.5
    money_left = balance - monthly_payment
    return monthly_payment, money_left
```

```python
pmt, money_left = loan_calculator(money_left, annual_rate, principal)
```

```python
print(pmt)
print(money_left)
```

6\. (8 pts) It is difficult to predict monthly car payments, but you may be in the market for a used or new car when graduating and starting at your new job. To simplify, create a function that takes two arguments `balance` and `car_cost`. There are many variable factors based on credit score, available financing, length of loan, but we can do a rough estimate using if-elif-else statements that follow the table below. Depending on the `car_cost` variable, return the `car_pmt` which is the monthly estimated car payment and `money_left` which is how much is left on the balance after subtracting the car payment.

Under 15000 - $250

15000-19999 - $300

20000-24999 - $400

25000-29999 - $500

30000-34999 - $600

35000-39999 - $700

40000 and over - $800

```python
cost = int(input("What is the cost of the car? $"))
```

```python
def car_pay(balance, car_cost):
    if car_cost < 15000:
        car_pmt = 250
    elif car_cost <= 19999:
        car_pmt = 300
    elif car_cost <= 24999:
        car_pmt = 400
    elif car_cost <= 29999:
        car_pmt = 500
    elif car_cost <= 34999:
        car_pmt = 600
    elif car_cost <= 39999:
        car_pmt = 700
    else:
        car_pmt = 800
    
    balance -= car_pmt
    return car_pmt, balance
```

```python
car_pmt, money_left = car_pay(money_left, cost)
print(car_pmt)
print(money_left)
```

7\. (8 pts) Like car payments, it is difficult to know how much your monthly housing costs will be, but they on average take up one-third of your budget. For this function, `housing` there will be 1 argument `balance` and 3 inputs the user gives: `home_cost`, `rate`, `years`. Use an if-else statement and compound condition to return "You should rent" or "You should buy". If the rate is greater than > 5% (.05) OR you plan on staying at that place for less than < 5 years then rent, otherwise buy. In addition, subtract the `home_cost` from balance and return that as `money_left`.  

```python
def housing(balance):
    home_cost = float(input("Home cost: $"))
    rate = float(input("Interest rate (as decimal): "))
    years = int(input("Years you plan to stay in the residence: "))
    money_left = balance - home_cost
    if rate > .05 or years < 5:
        str1 = "You should rent"
    else:
        str1 = "You should buy"
    return str1, money_left
```

```python
rent_or_buy, money_left = housing(money_left)
print(rent_or_buy)
print(money_left)
```
