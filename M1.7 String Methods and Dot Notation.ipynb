Using dot notation, you can list more than one methods to use. Dot notation is useful for taking the outcome of the previous method. That way, you do not have to store the intermediate outcome from the previous operation to do another operation. 

### String Methods

```python
s = "a"
s
```

```python
s.islower()
```

The `islower()` method returns True if the string is lowercase.

```python
s.isupper()
```

The `isupper()` method returns True if the string is uppercase.

```python
s = "This is text."
s
```

```python
s.upper()
```

The `upper()` method returns a string where all characters are in upper case. Symbols and numbers are ignored.

```python
s.lower()
```

The `lower()` method returns a string where all characters are lower case. Symbols and numbers are ignored.

```python
s.capitalize()
```

The `capitalize()` method is to return a copy of string with the first character capitalized and the rest lowercased. 

```python
s.count("t")
```

The `count` method returns the number of times a specified string appears in the string.

```python
s.lower().count("t")
```

All strings and string methods in Python are case-sensitive.

```python
s = "\tThis\nis\ntext.\n\n\n"     # \t: tab, \n: new line
print(s)
```

```python
print(s.strip('\n'))
```

The `strip()` method removes any leading (spaces at the beginning) and trailing (spaces at the end) characters. Space is the default character to be removed.

only leading and trailing found matches are removed; `strip('\n')` will only remove newline characters from the ends and not from anywhere in the middle of the string.

```python
s = ' \n \t hello\n'
```

```python
print(s)
```

```python
print(s.strip('\n'))
```

```python
print(s.lstrip())
```

The `lstrip()` method removes any leading characters.

```python
print(s.rstrip())
```

The `rstrip()` method removes any trailing characters.

```python
s = "This is text."
print(s.rstrip("."))
```

You can specify the character to strip out. 

```python
s.rstrip("t.")
```

```python
'www.example.com'.strip('cmow.')
```

The chars argument is a string specifying the set of characters to be removed. So all occurrences of these characters are removed, and not the particular given string.

```python
s = 'Arthur: three!'
s.lstrip('Arthur: ')
```

```python
s.removeprefix('Arthur: ')
```

```python
s.removesuffix('three!')
```

If we just want to remove the given string, we can use  `removeprefix()` and `removesuffix()`.

```python
s = "This is text."
s.startswith("This")
```

The `startswith()` method returns True if the string starts with the specified string, otherwise False.

```python
s.endswith("?")
```

The `endswith()` method returns True if the string ends with the specified string, otherwise False.

```python
s = "This is text."
s.find("text")
```

The `find()` method finds the index of the first occurrence of the specified string.

```python
s.find("z")
```

It returns -1 if the string is not found.

```python
s.index("text")
```

The `index()` method is almost the same as the `find()` method, the only difference is that the `find()`thod returns -1 if the string is not found.

```python
s.index("z")
```

```python
s = "This is text."
s.replace(" ", "-")
```

The `replace()` method replaces a specified string with another specified string.

```python
s = "This is text."
s.split()
```

The `split()` method splits a string into a list. Default separator is any whitespace. You can specify the separator. 

```python
s = "This_is_text."
s.split("_")
```

```python
l = ["This", "is", "text"]
" ".join(l)
```

The `split()` and `join()` methods are opposites of each other. The `join()` method takes all items in a list and joins them into one string. A string must be specified as the separator.

```python
"-".join(l)
```

```python
name = "Henry Tippie"
age = 95
```

```python
s = "Name: {}, Age: {}".format(name, age)
s
```

The `format()` method formats specified values in a string. A pair of matching curly braces is used as a placeholder, which will later be replaced by the value of a variable. 

```python
print("My name is {1} and I lived to be {0} years old.".format(95, "Henry Tippie"))
```

Note that string methods return a copy of a string, not changing the target string.
