## Exercises for For Loops

1\. Write a <b>for</b> loop that prints all integers from 1 to 100, separated by a tab. Use the <b>range</b> function to generate numbers.

```python
# Your answer here - Fill the code below
for num in range(1, 101):
    print(num, end="\t")
```

2\. Write a <b>for</b> loop that prints all characters in the string `s` below, separated by a tab.

```python
s = "abcdefghijklmnopqrstuvwxyz"
```

```python
# Your answer here
for ch in s:
    print(ch, end = "\t")
# a   b   c   
```

3\. Write a <b>for</b> loop that prints the same as above, but this time using the index of `s`.

```python
# Your answer here
for i in range(len(s)):
    print(s[i], end = "\t")
```

4. Write a <b>for</b> loop that prints the same as above, except that the letters at odd index positions are printed in upper case, while the other letters are printed as they are.

```python
# Your answer here
for i in range(len(s)):
    if i % 2 == 1: #Take i / 2, if the remainder is 1(i.e. the numerator is an odd number)
        print(s[i].upper(), end = "\t") #print the index[i] in upper case
    else:
        print(s[i], end = "\t") #print the index in lower case
        
```

## Exercises for While Loops

1\. Write a <code>`while`</code> loop to copy the strings <code>'orange'</code> of the list <code>squares</code> to the list <code>new_squares</code>. Stop and exit the loop if the value on the list is not <code>'orange'</code>:

```python
squares = ['orange', 'orange', 'orange', 'red', 'blue ', 'orange']
new_squares = []
```

```python
# Your answer here
i = 0
current_color = squares[0]

while current_color == "orange":
    new_squares.append(current_color)
    i += 1
    current_color = squares[i]
new_squares
```

2\. Write a <code>`while`</code> loop to print the first n natural numbers, where n is an integer that user inputs. hint: use the input() function to prompts user to enter a number.

```python
# Your answer here
n = int(input("Provide an integer"))
i = 0
while i <= n:
    print(i)
    i += 1
```

## Exercises for List Comprehension

1\. Not using and using list comprehension, create a list `l2` of squares of the numbers in `l1`.

```python
l1 = [1, 2, 3, 4, 5]
```

```python
# Your answer here NOT using list comprehension
l2 = []

for num in l1:
    l2.append(num **2)
print(l2)
```

```python
# Your answer here USING list comprehension
l2 = [num ** 2 for num in l1]
print(l2)
```

2\. Not using and using  list comprehension, create a list `l3` of lists, each of which contains the square and cube of each number in `l1`. For example, the last element in `l3` is [25, 125].

```python
# Your answer here NOT using list comprehension
l3 = []

for num in l1:
    l3.append([num **2, num **3])
print(l3)
```

```python
# Your answer here using list comprehension
l3 = [[num**2, num**3] for num in l1]
print(l3)
```

3\. Using list comprehension, create a list `names2` of the names in `names` in lower case.

```python
names = ["Alice", "Bob", "Tom"]
```

```python
# Your answer here
names2 = [name.lower() for name in names]
print(names2)
```

4\. Using list comprehension, create a list `names3` of tuples, each of which contains each name in `names` and the length of the name. For example, the first element in `names` is ('Alice', 5). 

```python
# Your answer here
names3 = [(name, len(name)) for name in names]
print(names3)
```

5\. Given a matrix (a list of lists) of numbers, use a list comprehension to create a new flattened list that contains only the even numbers from the matrix.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# expected output: [2, 4, 6, 8]
```

```python
# Your answer here

flatlist = [num for row in matrix for num in row if num %2 == 0] #For every row in matrix, for every number in row, if number is divisible by 2, accept number as an value in [flatlist]
print(flatlist)
```

## Exercises for Loop Control

1\. Write a <b>for</b> loop that prints all names in `names` except for "*Mary*". (Use <b>continue</b>.)

```python
names = ["Alice", "Bob", "Mary", "Sam", "Sarah", "Tom"]
```

```python
# Your answer here

for name in names:
    if name == "Mary":
        continue
    print(name)
```

2\. Write a <b>for</b> loop that prints the names in `names` from the beginning but stops right after it has printed "*Mary*".

```python
# Your answer here

for name in names:
    print(name)
    if name == "Mary":
        break
```

3\. Write a <b>for</b> loop that adds all numbers in `l` except for multiples of 5. Print the final sum after the <b>for</b> loop. (Use <b>continue</b>.)

```python
l = range(1, 1001)
```

```python
# Your answer here
total = 0

for i in l:
    if i % 5 == 0:
        continue
    total += i

print(total)
```

4\. Write a <b>for</b> loop that adds numbers in `l` from the beginning but stops when the intermediate sum exceeds 1000, printing the current sum and the number last added.

```python
# Your answer here

total = 0

for num in l:
    total += num
    if total > 1000:
        print("Current sum:", total)
        print("Last snumber added:", num)
        break
```
