```python
#For loop over collection[list]
list6 = []
for i in [50,40,30,20]:
    answer = i/2
    list6.append(i)
    list6.append(answer)
# x = 40
# answer = x/2
# list6.append(x)
# list6.append(answer)
# x = 30
# answer = x/2
# list6.append(x)
# list6.append(answer)
# x = 20
# answer = x/2
# list6.append(x)
# list6.append(answer)
print(list6)
#Output: [50, 25.0, 40, 20.0, 30, 15.0, 20, 10.0]
```

```python
#While loop stopping when variable reaches a certain number
i = 50
list6 = []
while i >= 20:
    answer = i/2
    list6.append(i)
    list6.append(answer)
    i -= 10
print(list6)
```

```python
#For loop in range
list6 = []
for i in range(50,10,-10):
    answer = i/2
    list6.append(i)
    list6.append(answer)
print(list6)
```

```python
#Nested loops
list6 = []
for num1 in range(50,10,-10):
    for num2 in range(25,5,-5):
        list6.append((num1, num2))
print(list6)
```

```python
#List comprehension, complete [] in one line
list6 = [(num1, num1/2) for num1 in range(50,10,-10)]
print(list6)
```

```python
#List comprehension, complete [] in one line
list6 = [item for num1 in range(50,10,-10) for item in [num1, num1/2]]
print(list6)
len(list6)
```

```python
#Continue (skip 1st element) and break before last element
i = 0
list6 = [50, 25.0, 40, 20.0, 30, 15.0, 20, 10.0]
newlist = []

while i < len(list6):
    if i == 0:
        i += 1
        continue
    if i == len(list6) - 1:
        break
        
    newlist.append(list6[i])
    i += 1

print(newlist)
 
#output should be 25.0, 40, 20.0, 30, 15.0, 20
```
