### Nested For Loops

```python
letters = ["letter1", "letter2", "letter3"]
steps = ["Fold the letter", "Insert the letter into an envelope", "Seal the envelope"]
```

```python
for letter in letters:
    print("Fold the letter")
    print("Insert it into an envelope")
    print("Seal the evelope")
    print
```

```python
for l in letters:               # outer for loop
    for s in steps:             # inner for loop
        print("Now I'm working on {}, with the step {}".format(l, s))
```

The inner <b>for</b> loop runs for every item in the outer **for** loop. 

```python
l1 = ['a', 'b', 'c']
l2 = ['1', '2', '3']
```

```python
for item1 in l1:                # outer for loop
    for item2 in l2:            # inner for loop
        print(item1 + item2)
```

```python
for item2 in l2:               # outer for loop
    for item1 in l1:           # inner for loop
        print(item2 + item1)
```

The order of retrieving elements depends on how it is defined in the nested <b>for</b> loop.

```python
l1 = ["a", "b", "c"]
l2 = ["1", "2", "3", "4"]
l3 = ["p", "q", "r"]

count = 0
for item1 in l1:
    for item2 in l2:
        for item3 in l3:
            print(item1 + item2 + item3)
            count += 1
print(count)
```

You can write a nested <b>for</b> loop with a depth of 3 or more. 

```python
l = ["a", "b", "c"]

for item1 in l:
    for item2 in l:
        for item3 in l:
            print(item1 + item2 + item3)
```

You can write a nested <b>for</b> loop even using a single list.
