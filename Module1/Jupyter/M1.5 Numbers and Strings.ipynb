### String Additions

````python
s1 = "How are you "
s2 = "world?"

s1 + s2
````

The easiest way to combine two strings is to use the + operator. 

````python
s1 + " " + s2
````

````python
s1 * 3
````

````python
(s1 +  " " + s2 + " ")*3
````

### String Containment

````python
s1 = "Python"
s2 = "Py"
s3 = "Pie"

s2 in s1
````

The ```in``` operator returns True if the first operand is contained in the second.
- Note that True and False are Boolean data types that will be covered in Module 2
