```python

# Example 1: Data Visualization
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

```python

# Example 2: Data Cleaning
# Removing None values from a list
raw_data = [10, None, 25, None, 40]
cleaned_data = [x for x in raw_data if x is not None]
print("Cleaned Data:", cleaned_data)
```

```python

# Example 3: Basic Analysis
# Calculating average from a list of numbers
scores = [85, 90, 78, 92, 88]
average_score = sum(scores) / len(scores)
print("Average Score:", average_score)
```
