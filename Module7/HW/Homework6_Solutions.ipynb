2\.	Transform your mpg column into a new categorical variable mpg_level. If it is greater than the median of the column mpg_level = 1, otherwise mpg_level = 0.

3\.	Using sklearn, split your dataset into an 80/20 training/testing split

```python
#Answer here
X = dfc.drop(['mpg','name','origin','mpg_level'], axis=1)
y = dfc['mpg_level']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

## 4\. Perform the following 3 classification techniques using mpg_level as your target (y) variable and all the rest as your predictor (X) variables (do not include name and original mpg in your X variables).
###	Logistic Regression
###	KNN with neighbors = 7
###	Decision Tree


```python
#Answer here

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Logistic Regression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
log_preds = log_model.predict(X_test)
log_acc = log_model.score(X_test, y_test)

# KNN Classifier
knn_model = KNeighborsClassifier(n_neighbors=7)
knn_model.fit(X_train, y_train)
knn_preds = knn_model.predict(X_test)
knn_acc = knn_model.score(X_test, y_test)

# Decision Tree Classifier
tree_model = DecisionTreeClassifier(random_state=1)
tree_model.fit(X_train, y_train)
tree_preds = tree_model.predict(X_test)
tree_acc = tree_model.score(X_test, y_test)

# Print accuracy scores
print("Logistic Regression Accuracy:", log_acc)
print("KNN Accuracy (k=7):", knn_acc)
print("Decision Tree Accuracy:", tree_acc)

```
