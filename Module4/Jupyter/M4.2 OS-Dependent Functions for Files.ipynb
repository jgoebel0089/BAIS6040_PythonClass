## Useful Operating System-Dependent Functions for Files

```python
import os
```

The <b>os</b> module provides a portable way of using operating system dependent functionality for handling files and directories.

os dcumentations:
https://docs.python.org/3/library/os.html#files-and-directories

```python
os.listdir(path=".")                        # The path "." means the current directory.
```

The <b>os.listdir</b>(path='.') function returns a list containing the names of the entries in the directory given by `path`. The list is in arbitrary order, and does not include the special entries

```python
[item for item in os.listdir() if item.endswith(".ipynb")]     # a list of ipynb files in the current directory
```

```python
os.getcwd()
```

The <b>os.getcwd</b> function returns a string representing the current working directory.

```python
os.path.isfile("C:/OneDrive - University of Iowa/BAIS_6040/Module 4/M4.2 OS-Dependent Functions for Files.ipynb")
```

The <b>os.path.isfile</b>(path) function returns True if `path` is an existing regular file. 

```python
os.path.isdir("../Module 4")        # ".." means the parent directory
```

The <b>os.path.isdir</b>(path) function returns True if `path` is an existing directory (folder).

```python
os.chdir('../New_Module')  # change the current directory to my own folder "New_Module"
```

```python
os.getcwd()                                # you can confirm by displaying the current work directory
```

The <b>os.chdir</b>(path) function changes the current working directory to path.

```python
if not os.path.isdir("outcome"):         # Check if there is an existing directory named outcome.
    os.mkdir("outcome")                  # Create a new directory named outcome.
```

The <b>os.mkdir</b>(path) function creates a directory named `path`.

We are going to save all of our outcome files in the `outcome` directory.
