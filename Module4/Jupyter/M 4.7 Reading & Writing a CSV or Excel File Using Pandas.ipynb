## Reading and Writing a CSV File Using Pandas

```python
import pandas as pd     
```

```python
df = pd.read_csv("outcome/my_titanic.csv")
df
```

pandas.read_csv: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html

When you need to read a CSV file and analyze the content in a tabular format with rows and columns, it is a good idea to read the file into a Pandas dataframe. 

```python
df = pd.read_csv("outcome/my_titanic.csv", sep="\t")
df
```

```python
df = pd.read_csv("https://people.sc.fsu.edu/~jburkardt/data/csv/biostats.csv")
df
```

## Reading Excel Tables with pandas 

```python
#! pip install --user openpyxl xlrd xlwt xlsxwriter 
```

If you need to read Excel spreadsheet into a Python program, you can use the pandas `read_excel()` function, which is similar to the `read_csv()` except instead of providing the name of a text file, we provide the name of an Excel file(.xls or .xlsx).

```python
df = pd.read_excel("example_spreadsheet.xlsx", sheet_name="Sheet2")
```

```python
df
```

pandas.read_excel: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html

## Write pandas DataFrame to CSV File and Excel

```python
df.to_csv("outcome/my_data.csv", sep=",", index=False)
```

pandas.DataFrame.to_csv: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html

```python
df.to_excel("outcome/my_data.xlsx", sheet_name="SALES1", index=False)
```

pandas.DataFrame.to_excel: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_excel.html
