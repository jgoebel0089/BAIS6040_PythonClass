# Notebook Demonstration 4.8 - Reading & Writing a JSON File

JSON, which stands for JavaScript Object Notation, is one of the most commonly used formats for data transfer. It is preferred, because it is clean, easy to read, and easy to parse. Many websites provide JSON-enabled APIs, or Application Programming Interfaces. 

```python
import json
import pandas as pd
```

```python
states = {"IL": "Illinois", "WI": "Wisconsin", "IA": "Iowa", "NE": "Nebraska", "MN": "Minnesota"}
states
```

```python
with open("outcome/my_states.json", "w") as fw:
    json.dump(states, fw)
```

The <b>json.dump</b>(obj, fp, ...) function serializes `obj` as a JSON formatted stream to `fp`(file-like object).

https://docs.python.org/3/library/json.html#json.dump

Writing a Python object to a file is called serialization. If you manually open the JSON file and see the content, you will see that it looks just like a Python dictionary. 

```python
with open("outcome/my_states.json", "r") as fr:
    states_new = json.load(fr)
```

The <b>json.load</b>(fp, ...) deserializes `fp` to a Python object. 

https://docs.python.org/3/library/json.html#json.load

Reading a file back into a Python object is called deserialization. 

```python
states_new
```

### Reading a flat .json file into a Pandas data frame

```python
import pandas as pd

# use the read_json() method
df = pd.read_json("iris.json", encoding="utf-8")


print(df.head())
```

```python
# first, read in some data
import pandas as pd

df = pd.read_csv('mtcars.csv', encoding='utf-8', sep=',')

print(df)
```

```python
# write JSON file using utf-8 encoding

# first open a file in write mode and specify utf-8 encoding
with open("outcome/mtcars.json", 'w', encoding='utf-8') as file:
    # then write the data using Pandas .to_json()
    df.to_json(file, orient='records', lines=False)
```
