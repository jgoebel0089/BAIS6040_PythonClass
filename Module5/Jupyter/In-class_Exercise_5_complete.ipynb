# Pandas Dataframes - aggregating, working with missing values, feature engineering

```python
#Importing pandas as pd
import pandas as pd
```

```python
#Read in FAANG csv file below using pd.read_csv() as df_stock
# "Q1_FAANG_Final.csv"
df_stock = pd.read_csv("Q1_FAANG_Final.csv")
```

### Create a new column called Daily_Change that looks at the percent change from Open => Close. This formula is (Open - close)/Open

```python
df_stock['Daily_Change'] = (df_stock['Opening_Price'] - df_stock['Closing_Price']) / df_stock['Opening_Price']
```

```python
df_stock
```

### Use .loc in order to just get the Date, Ticker, and Daily_Change columns. Save this as df_copy

```python
df_copy = df_stock.loc[1:10,['Date','Ticker','Daily_Change']] #Rows 1-10, : All rows
df_copy
```

### With df_copy use dataframe filtering to get only rows that have Netflix as their stock. Save this as df_net

```python
#toms = actors_df[ actors_df['First Name'] == 'Tom' ]  See example for help. Instead of First Name it'll be Ticker
df_net = df_copy[ df_copy['Ticker'] == 'NFLX']
df_net
```

```python
df_net = df_copy[ df_copy.Ticker == 'NFLX']
df_net
```

### Use .agg in order to just get the min, max, and mean for Daily_Change columns for df_net

```python
#actors_df['Age'].agg(['min', 'max', 'mean'])   See example here for help
df_net['Daily_Change'].agg(['min', 'max', 'mean'])
```

# Taylor Swift Spotify

```python
#Read in taylor_swift_spotify.csv file using pd. Use df_ts
df_ts = pd.read_csv("taylor_swift_spotify.csv")
df_ts
```

### Groupby album and get the mean values for each column using .mean(numeric_only=True). Save this as df_album

```python
df_album = df_ts.groupby('album').mean(numeric_only=True)
```

```python
df_album['duration_min'] = df_album['duration_ms']/60000
```

```python
df_album.sort_values('duration_min')
```

```python
df_album
```

# Office Employees File

```python
import pandas as pd
df_office = pd.read_csv("office_employees.csv")

print(df_office)
```

```python
#Use .duplicated to find the duplicate records of office employees. How many are there?
duplicates = [df_office.duplicated(keep=False)]
duplicates
```

```python
#Use .drop_duplicates to drop them.
df_off = df_office.drop_duplicates()
```

```python
# Change the age column to numeric using df_off['Age'] = pd.to_numeric(df_off['Age'], errors='coerce')
df_off['Age'] = pd.to_numeric(df_off['Age'], errors='coerce')
```

### Find which row has NaN using .isna.  Use .fillna() to make that missing value 28.

```python
df_off['Age'].isna()
```

```python
df_off['Age'].fillna(28)
```
