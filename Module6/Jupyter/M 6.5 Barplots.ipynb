```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Reviews_update.csv", index_col=0)
```

### Barplot of Scores: 

```python
# Count the frequency of each score
score_counts = df.Score.value_counts().sort_index()

# Create the bar plot
plt.bar(score_counts.index, score_counts.values, color='gray', align="center", width=0.5, edgecolor="red", linewidth=1)

# Add a title
plt.title('Frequency of Review Scores')


# Add labels
plt.xlabel('Review Score')
plt.ylabel('Number of Reviews')

# Display the figure
plt.show()
```

```python
df = pd.read_csv("taylor_swift_spotify.csv")
```

```python
# Bar plot with multiple categories 

df_bar = df.groupby('album')[['popularity', 'energy', 'danceability']].mean().sort_values('popularity', ascending = False)

x_vals = np.arange(df_bar.shape[0])
plt.bar(x_vals, height = df_bar['popularity']/100, width = 0.25, label = 'popularity')
plt.bar(x_vals+0.25, height = df_bar['energy'], width = 0.25, label = 'energy')
plt.bar(x_vals+0.5, height = df_bar['danceability'], width = 0.25, label = 'danceability')
plt.ylabel('Average value')
plt.xticks(ticks = x_vals+0.25, labels = df_bar.index, rotation='vertical')
plt.title('Average popularity, energy and danceability by album')
plt.legend()
plt.show()
```

```python
# Bar plot with multiple categories 

df_bar = df.groupby('album')[['popularity', 'energy', 'danceability']].mean().sort_values('popularity', ascending = False)

x_vals = np.arange(df_bar.shape[0])

fig, ax = plt.subplots(figsize = (10,4))

ax.bar(x_vals, height = df_bar['popularity']/100, width = 0.25, label = 'popularity')
ax.bar(x_vals+0.25, height = df_bar['energy'], width = 0.25, label = 'energy')
ax.bar(x_vals+0.5, height = df_bar['danceability'], width = 0.25, label = 'danceability')
ax.set_ylabel('Average value')
ax.set_xticks(ticks = x_vals+0.25, labels = df_bar.index, rotation='vertical')
ax.set_title('Average popularity, energy and danceability by album')
ax.legend()
plt.show()
```
