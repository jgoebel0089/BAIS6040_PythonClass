# Notebook Demonstration 6.4 Scatterplots and Line Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("taylor_swift_spotify.csv")
```

```python
df.head()
```

# Scatterplot

```python
# Draw a scatter plot of energy vs loudness

# Using pyplot interface:

plt.scatter(df['loudness'], df['energy'])
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.title('Energy vs. Loudness')
plt.show()
```

```python
# Draw a scatter plot of energy vs loudness

# Using figure and axes objects
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(df['loudness'], df['energy'])
ax.set_xlabel('Loudness')
ax.set_ylabel('Energy')
ax.set_title('Energy vs. Loudness')
plt.show()
```

```python
# Same as before, but let's set the marker color to red for the album 'Red'

colors = ['red' if a == 'Red' else 'blue' for a in df['album'] ]
plt.scatter(df['loudness'], df['energy'], color = colors)
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

```python
# Same as before, but let's set the marker color to orange for the album 'Red'

colors = ['orange' if a == 'Red' else 'blue' for a in df['album'] ]
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(df['loudness'], df['energy'], color = colors)
ax.set_xlabel('Loudness')
ax.set_ylabel('Energy')
plt.show()
```

# Line graph

```python
df["release_date"] = pd.to_datetime(df['release_date'])
df["year"] = df["release_date"].dt.year
df
```

```python
# 1. Compute average popularity for each year
# 2. Draw a line plot to display the average values over years

df_year = df.groupby('year')[['popularity']].mean()

plt.plot(df_year.index, df_year['popularity'])
plt.xlabel('Year')
plt.ylabel('Average popularity')
plt.title('Average popularity by year')
plt.grid(axis='y')
plt.xticks(ticks = df_year.index, labels = df_year.index, rotation=60)
plt.show()
```
