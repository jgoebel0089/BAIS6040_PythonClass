# Scatterplot

```python
# Draw a scatter plot of energy vs loudness

# Using pyplot interface:

plt.scatter(df['loudness'], df['energy'])
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.title('Energy vs. Loudness')
plt.show()
```

```python
# Draw a scatter plot of energy vs loudness

# Using figure and axes objects
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(df['loudness'], df['energy'])
ax.set_xlabel('Loudness')
ax.set_ylabel('Energy')
ax.set_title('Energy vs. Loudness')
plt.show()
```

```python
# Same as before, but let's set the marker color to red for the album 'Red'

colors = ['red' if a == 'Red' else 'blue' for a in df['album'] ]
plt.scatter(df['loudness'], df['energy'], color = colors)
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

```python
# Same as before, but let's set the marker color to orange for the album 'Red'

colors = ['orange' if a == 'Red' else 'blue' for a in df['album'] ]
fig, ax = plt.subplots(figsize = (5,5))
ax.scatter(df['loudness'], df['energy'], color = colors)
ax.set_xlabel('Loudness')
ax.set_ylabel('Energy')
plt.show()
```

# Line graph

```python
df["release_date"] = pd.to_datetime(df['release_date'])
df["year"] = df["release_date"].dt.year
df
```

```python
# 1. Compute average popularity for each year
# 2. Draw a line plot to display the average values over years

df_year = df.groupby('year')[['popularity']].mean()

plt.plot(df_year.index, df_year['popularity'])
plt.xlabel('Year')
plt.ylabel('Average popularity')
plt.title('Average popularity by year')
plt.grid(axis='y')
plt.xticks(ticks = df_year.index, labels = df_year.index, rotation=30)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Compute average popularity for each year
df_year = df.groupby('year', as_index=False)['popularity'].mean()

# 2. Draw a Seaborn lineplot
sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(10,6))
sns.lineplot(
    data=df_year,
    x="year",
    y="popularity",
    marker="o",
    linewidth=2.5,
    color="royalblue"
)
# Labels & title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average popularity", fontsize=14)
plt.title("Average popularity by year", fontsize=16, weight="bold")

# Rotate ticks for readability
plt.xticks(rotation=30)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# 1. Compute average popularity for each year
df_year = df.groupby('year', as_index=False)['popularity'].mean()

# 2. Draw a Seaborn lineplot
sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(10,6))
sns.lineplot(
    data=df_year,
    x="year",
    y="popularity",
    marker="o",
    linewidth=2.5,
    color="royalblue"
)

# Labels & title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average popularity", fontsize=14)
plt.title("Average popularity by year", fontsize=16, weight="bold")

# Force integer ticks
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.xticks(rotation=30)

plt.show()
```

# Expected product from Copilot

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("taylor_swift_spotify.csv")

# Convert release_date to datetime and extract year
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['year'] = df['release_date'].dt.year

# Group by year and calculate mean popularity
popularity_by_year = df.groupby('year')['popularity'].mean().reset_index()

# Set seaborn style
sns.set_style("whitegrid")

# Create the line plot
plt.figure(figsize=(14, 7))
plt.plot(popularity_by_year['year'], popularity_by_year['popularity'], marker='o', linestyle='-')
plt.title("Average Popularity of Taylor Swift Songs by Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Mean Popularity", fontsize=14)
plt.ylim(35, 80)
plt.xticks(ticks=popularity_by_year['year'], labels=popularity_by_year['year'].astype(int), rotation=30)
plt.grid(True)
plt.tight_layout()

# Save the plot
plt.savefig("mean_popularity_by_year_seaborn_fixed.png")
```
