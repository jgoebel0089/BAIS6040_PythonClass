```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Histograms (distribution of the polarity scores)

histograms: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
df = pd.read_csv("Reviews_update.csv", index_col=0)
```

```python
plt.hist(df.compound_score)
plt.show()
```

```python
plt.hist(df.compound_score, edgecolor='black')

plt.ylabel("frequency")
plt.xlabel("compound_score")
plt.title("Histogram Plot of the Compound Scores")
plt.show()
```

```python
heights = np.random.normal(160, 5, 1000)
plt.hist(heights, bins=15)
plt.show()
```

```python
# cumulative histogram
heights = np.random.normal(160, 5, 1000)
plt.hist(heights, bins=15 , cumulative=True)
plt.show()
```

###  Boxplot for Sentiment Scores

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
plt.boxplot(df[['neg_score', 'neutral_score', 'pos_score', 'compound_score']], 
            labels=['Negative', 'Neutral', 'Positive', 'Compound'])

plt.title('Boxplot of Sentiment Scores')
plt.ylabel('Scores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("taylor_swift_spotify.csv")
df['duration'] = df['duration_ms']/1000
```

```python
# Side by Side Histograms

# Plot histograms of duration and popularity side-by-side

# Using figure and axes objects
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (10,3))
axs[0].hist(df['duration'])
axs[0].set_xlabel('Duration (seconds)')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Distribution of duration')

axs[1].hist(df['popularity'], color = 'magenta')
axs[1].set_xlabel('Popularity')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Distribution of popularity')

plt.show()
```
