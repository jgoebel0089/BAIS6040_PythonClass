## Matplotlib basics

The structure of Matplotlib. It consists of three main layers:
- backend layer: 
contains three abstract classes: 1. FigureCanvas(get a blank paper to draw); 2.Renderer(what tool we use); 3.Event(what to draw)
- artist layer:
contains one object - Artist. (Best for the programmers, but a bit heavy for the everyday purposes) 
- scripting layer: 
a lighter interface than Artist layer. (matplotlib.pyplot, our focus)

```python
import matplotlib.pyplot as plt
```

<b>matplotlib.pyplot</b> is a collection of functions that make matplotlib work like MATLAB: https://matplotlib.org/stable/api/pyplot_summary.html#module-matplotlib.pyplot

Generating visualizations with pyplot is very quick:

```python
plt.plot(11, 6, "o")
# plt.axis((0, 15, 0, 15)) # to set axis properties (xmin, xmax, ymin, ymax)
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(0., 5., 0.1)
y1 = x
y2 = np.exp(x)

plt.plot(x, y1, "r.")   # red points  # change the markersize: markersize=3
plt.plot(x, y2, "g^")   # green triangles
plt.axis((0, 5, 0, 150))
plt.show()
```

- pandas.Series.plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.html
- pandas.DataFrame.plot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html


When plotting a plot on a Pandas series using the <b>plot</b> method, the x axis is the index of the series, and the y axis is its values.

`kind`: str
- line: line plot (default)
- bar: vertical bar plot
- barh: horizontal bar plot
- hist: histogram
- scatter: scatter plot (DataFrame only)
- box: boxplot
- kde: Kernel Density Estimation plot
- density: same as ‘kde’
- area: area plot
- pie: pie plot
- hexbin: hexibin plot (DataFrame only)

When plotting a plot on a Pandas dataframe using the <b>plot</b> method, the x axis is the index of the dataframe, and the y axis is the values of the columns in the dataframe.

plot types: https://matplotlib.org/stable/plot_types/index.html
