3\. There are 7 unique colors in the `color` column of `dfdi`: *D*, *E*, *F*, *G*, *H*, *I*, and *J*. Plot a pie chart and a bar plot in one figure, two columns in one row. The pie chart displays the distribution(percentage) of the count of each color and the bar plot displays the mean prices for the 7 colors. Set the `figsize` to 14 x 7. Use the <b>groupby</b> method to calculate the mean prices for the 7 colors. 

```python
# Ensure we use the correct column name (lowercase)
# Compute counts for the 7 colors, ordered D..J
expected_colors = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
color_counts = (dfdi['color'].value_counts().reindex(expected_colors, fill_value=0))
```

```python
# Your answer here
fig, axs = plt.subplots(nrows = 2, ncols = 1, figsize = (14,7))


# --- Pie chart: distribution (percentages) ---
axs[0].pie(
    x=color_counts.values,            # numeric sizes
    labels=color_counts.index,        # labels per wedge
    autopct='%1.1f%%',                # show percentages
    startangle=90,                    # rotate to start at the top
    counterclock=False
)
axs[0].axis('equal')                  # make the pie circular
axs[0].set_title('Color Distribution (Percentage of Diamonds)', fontsize=14, pad=12)



# --- Bar plot in the second axes: mean price by color ---
mean_prices = (dfdi.dropna(subset=['price'])
               .groupby('color')['price']
               .mean()
               .reindex(expected_colors))

axs[1].bar(mean_prices.index, mean_prices.values, color='#5DADE2', edgecolor='#1F618D')
axs[1].set_title('Mean Price by Color (USD)', fontsize=14, pad=12)
axs[1].set_xlabel('Color', fontsize=12)
axs[1].set_ylabel('Mean Price (USD)', fontsize=12)
axs[1].grid(True, axis='y', linestyle='--', alpha=0.35)

plt.tight_layout()
plt.show()
```

5\. Plot a Pie Chart to show the total number of passergers survived or not (died). Set the `startangle` as 90, add the percentage label inside the pie with the format `%.1f%%`, and split the two parts by setting the `explode` as \[0, 0.1\], set the color as "green" for survived group and "gray" for non-survived group. 

```python
dft.head()
```

```python
# Your answer here

# Count survivors (1) and non-survivors (0)
counts = dft['survived'].value_counts().reindex([0, 1])

labels = ['Did not survive', 'Survived']
colors = ['gray', 'green']
explode = [0, 0.1]  # explode the 'Survived' slice



plt.pie(
    x=counts.values,
    labels=labels,
    colors=colors,
    explode=explode,
    startangle=90,
    autopct='%.1f%%',
    pctdistance=0.7
)
plt.show()
```

6\. Plot two pie charts, two columns in one row. Set the `figsize` to 14 x 7. The left pie chart shows the passenger class `Pclass` distribution for the survived passengers; the pie chart on the right shows the `Pclass` distribution for passengers who didn't survive. Add the text labels (given in the `labels` list) and the percentage label inside each pie chart with the format `%.1f%%`, and add a title for each pie chart.

```python
# Your answer here
# Fill your code below
fig, axs = plt.subplots(1, 2, figsize=(14,7))

ax1 = axs[0]
ax2 = axs[1]

survived = dft[dft['survived']==1]
not_survived = dft[dft['survived']==0]


# Labels for the three passenger classes in order of Pclass 1, 2, 3
labels = ['First Class', 'Second Class', 'Third Class']


# Compute class counts in fixed order [1, 2, 3] so labels align with slices
surv_counts = survived['pclass'].value_counts().reindex([1, 2, 3], fill_value=0)
not_surv_counts = not_survived['pclass'].value_counts().reindex([1, 2, 3], fill_value=0)



# Pie chart on the left is for passengers who survived

ax1.pie(
    surv_counts.values,
    labels=labels,
    colors=colors,
    autopct='%.1f%%',     # percentage label format
    startangle=90,        # start angle
    pctdistance=0.7       # place percentages inside the pie
)
ax1.axis('equal')         # keep the pie circular
ax1.set_title('Pclass Distribution — Survived', fontsize=14, pad=12)



# Pie chart on the right for passengers who died

ax2.pie(
    not_surv_counts.values,
    labels=labels,
    colors=colors,
    autopct='%.1f%%',
    startangle=90,
    pctdistance=0.7
)
ax2.axis('equal')
ax2.set_title('Pclass Distribution — Did Not Survive', fontsize=14, pad=12)



# Show the plot

plt.tight_layout()
plt.show()
```
