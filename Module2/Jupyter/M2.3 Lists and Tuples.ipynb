## Lists

A list is a collection that is ordered, mutable, indexed, and written with square brackets. It allows duplicate members.

```python
list1 = []
print(list1, type(list1))
```

```python
list1 = [1, 2, 3]
list1
```

You can create a list by listing its elements in a pair of matching square brackets. 

```python
l = ["a", "b", "c"]
l
```

```python
l = [1, 2, 3, "a", "b", "c"]
l
```

List elements do not have to be of the same type.

```python
len(l)
```

The length of a list is the number of items in the list.

```python
l[:3]
```

Indexing and slicing of lists are exactly the same as that of strings.   ***[start:stop:step]***

```python
l[-3:]
```

```python
l[::-1]
```

```python
3 in l
```

```python
"z" in l
```

```python
l1 = [1, 2, 3]
l2 = ["a", "b", "c"]

l1 + l2
```

```python
l1*2 + l2*3
```

The easiest way to combine two lists is to use the + operator, just as combining two strings. 

Note that `l1` + `l2` returns a new combined list, not changing either `l1` or `l2`.

### List Methods

```python
l1.extend(l2)
l1
```

The `extend()` method extends the list by appending all the items from another list. Note that the `extend()` method actually extends the target list, not returning a copy.

```python
l1 = [1, 2, 3]
l2 = ["a", "b", "c"]

l1 = l1 + l2
l1
```

Another way to extend a list is to combine the target list with the new list using the + operator and then store the outcome back in the target list.  

```python
l = [1, 2, 3]

l.append(4)
l
```

The `append` method appends an element to the end of the list. Note that the `append` method actually appends a new element to the target list, not returning a copy. 

```python
l.append([5, 6])
l
```

The list [5,6] has been added as an element of `l`, not part of the list. The `append` method always adds one element to the end of a list.

```python
fruits = ['apple', 'banana', 'cherry']
fruits.insert(2, "orange")
fruits
```

The `insert(*pos, *element)` method inserts the specified value `*element` at the specified position `*pos`.

```python
l = [1, 2, 3, 4, 5]
l.pop()
l
```

The `pop(*pos)` method removes the element at the specified position. The default value of `*pos` is -1, which removes the last item.

```python
pop1 = l.pop(0)
l, pop1
```

`pop(0)` will remove the first item.

```python
l.remove(4)
l
```

The `remove` method removes the __first__ occurrence of the element with the specified value.

```python
l = [1, 6, 3, 4, 2, 5]
l.sort()
l
```

The `sort()` method sorts the list in ascending order by default.

```python
l = [1, 6, 3, 4, 2, 5]
l.sort(reverse=True)
l
```

Setting the `reverse` parameter to True will sort the list in descending order.

```python
l = [1, 6, 3, 4, 2, 5]
l.reverse()
l
```

The `reverse` method reverses the sorting order of the elements.

```python
l[::-1]
```

Another way to get a reversed list is to use slicing setting the `step` parameter to -1. 

```python
l = [1, 2, 3, 4, 5, 4, 3, 2, 1]
l.count(2)
```

The `count` method returns the number of elements with the specified value.

## Tuples

A tuple is a collection that is ordered, immutable, indexed, and written with round brackets. It allows duplicate members.

```python
t = ("a", "b", "c")
print(t, type(t))
```

You can create a tuple by listing its elements in a pair of matching round brackets. 

```python
t[0]
```

```python
t[-1]
```

Indexing and slicing of tuples are the same as other collections.

```python
t[0] = "A"
```

Once a tuple is created, you cannot change its values as tuples are immutable. 

In general,  we use a list to store similar, or homogeneous, items, whereas we use a tuple to store heterogeneous items describing an entity. 

```python
l = ['a','b','c']
l[0] = 'A'
l.append('d')
t = tuple(l)
```

```python
t
```

```python
employees = [("Alice", 30, "female"), ("Bob", 25, "male"), ("Tom", 34, "male")]
```

In this example, `employees` is a list of three tuples, each of which consists of three items.

```python
employees[0]
```

```python
employees[0][0]
```

```python
employees[0][1]
```
