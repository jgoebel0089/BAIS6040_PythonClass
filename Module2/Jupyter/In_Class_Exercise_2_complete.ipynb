# Branching

Write an <b>if-elif-else/b> statement that prints "*x is greater than y.*" if `x` is greater than `y`, "*x is less than y.*" if `x` is less than `y`, and "*x equals y.*" otherwise.  Test with different values of x and y. 

```python
# Your answer here
def logic(x,y):
    if x > y:
        print("x is greater than y")
    elif x < y:
        print("x is less than y")
    else:
        print("x equals y")
    return

logic(0,9)
```

Have the user input two weather variables: temp and precip. You will need to convert temp to a float and precip can be a "Yes" or "No" string.

* First check if there is no precipitation and the temperature is > 80, print("It is too hot!")
* Then, if that is false, check if the temperature is < 32 and there is precipitation, then print("Not dealing with snow!")
* Otherwise, print ("I can work with this weather!")

```python
# Your answer here
temp = float(input("What is the temp?"))
precip = str(input("Rain yes or no?"))
print(temp)
print(precip)

if temp > 80 and precip == "no":
    print("It is too hot")
elif temp < 32 and precip == "yes":
    print("Not dealing with the snow")
else: print("I can work with this weather")
```

# Functions and Branching

```python
# Write a function letter_grade(score) using if/elif/else logic inside
# A: 90+, B: 80–89, C: 70–79, D: <70

# Test the function on 3 scores: 75, 90, 80
def letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else: return "FAIL"

print(letter_grade(75))
print(letter_grade(90))
print(letter_grade(80))
```
