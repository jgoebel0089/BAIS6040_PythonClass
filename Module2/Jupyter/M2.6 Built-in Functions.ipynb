### Difference between Functions and Methods

- A function looks like this: `function_name(something)`, e.g., sorted(l1)
- A method looks like this: `something.method_name()`, e.g., l1.sort()

A method always belongs to an object (e.g. string methods only work for string objects, list methods only work for list objects, etc.), while a function doesn’t necessarily (e.g., you can use the `len` function for strings, lists, tuples, or any data types).
