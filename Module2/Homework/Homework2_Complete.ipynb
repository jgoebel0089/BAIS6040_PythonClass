3\. (8 pts) Define a function `roth_ira` that takes one argument `balance`, which is your current balance before investing in your monthly Roth IRA retirement. 5\. (8 pts) You will be creating a loan_calculator function that calculates the student loan monthly payment shown at this website. We will assume a standard repayment plan of 10 years and 120 payments.

https://lendedu.com/blog/student-loan-payment-calculator/

Outside of the function, get the annual rate (.07 for 7 percent) and principal.
When defining the function, you have 3 arguments. The first argument is your `balance`, which is the same as Problem 3 or what is left in your monthly budget. The other 2 arguments are `rate` and `loan_amt`, which we got as input from the user in the cell before. The default arguments for those will be .075 and 20000.

Inside of the function, calculate the monthly loan payment based on the calculation from the website above. Return `monthly_payment` and `money_left` on your budget after you pay your student loan monthly payment.

```python
annual_rate = float(input("What is the annual rate?")) #Set as 7.5%(0.075)
principal = float(input("What is the principal amount")) #Set as $20000
```

```python
def loan_calculator(money_after_roth, annual_rate, principal): #Student loan calculator function
    monthly_payment = principal * (annual_rate/365.25) * (1+(annual_rate/365.25))**3652.5 / (((1+annual_rate/365.25)**3652.5)-1) * 30.5 #Determine student loan monthly payment
    balance = money_after_roth - monthly_payment #Determine remaining balance
    return monthly_payment, balance #Return student loan monthly payment and the remaining balance after student loan payment is deducted
```

```python
monthly_payment, money_after_studentloans = loan_calculator(money_after_roth, annual_rate, principal)
print(f"Monthly payment: ${monthly_payment:.2f}")
print(f"Balance after Roth IRA and Student Loan payment: ${money_after_studentloans:.2f}")
```

6\. (6 pts) It is difficult to predict monthly car payments, but you may be in the market for a used or new car when graduating and starting at your new job. To simplify, create a function that takes two arguments `balance` and `car_cost`. There are many variable factors based on credit score, available financing, length of loan, but we can do a rough estimate using if-elif-else statements that follow the table below. Depending on the `car_cost` variable, return the `car_pmt` which is the monthly estimated car payment and `money_left` which is how much is left on the balance after subtracting the car payment.

Under 15000 - $250

15000-19999 - $300

20000-24999 - $400

25000-29999 - $500

30000-34999 - $600

35000-39999 - $700

40000 and over - $800

```python
def car_pay(money_after_studentloans, car_cost): #Monthly car payment based on car value
    if car_cost < 15000:
        car_pmt = 250
    elif 15000 <= car_cost <= 19999:
        car_pmt = 300
    elif 20000 <= car_cost <= 24999:
        car_pmt = 400
    elif 25000 <= car_cost <= 29999:
        car_pmt = 500
    elif 30000 <= car_cost <= 34999:
        car_pmt = 600
    elif 35000 <= car_cost <= 39999:
        car_pmt = 700
    else: car_pmt = 800

    money_after_carloan = money_after_studentloans - car_pmt #Calculate remaining balance after monthly car payment deducted
    return car_pmt, money_after_carloan #Return monthly car payment and remaining balance
```

```python
car_cost = float(input("What is the cost of the car?")) #What is price of car?
monthly_car, money_after_carloan = car_pay(money_after_studentloans, car_cost) #Execute car payment function
print(f"The monthly car payment will be: ${monthly_car}")
print(f"Balance after Roth IRA, Student Loan payment, and Car Loan payment: ${money_after_carloan:.2f}")
```

7\. (6 pts) Like car payments, it is difficult to know how much your monthly housing costs will be, but they on average take up one-third of your budget. For this function, `housing` there will be 1 argument `balance` and 3 inputs the user gives: `home_cost`, `rate`, `years`. Use an if-else statement and compound condition to return "You should rent" or "You should buy". If the rate is greater than > 5% (.05) OR you plan on staying at that place for less than < 5 years then rent, otherwise buy. In addition, subtract the `home_cost` from balance and return that as `money_left`.  

```python
def housing(money_after_carloan): #Monthly rent/mortgage function
    home_cost = float(input("What is the monthly home price?"))
    rate = float(input("What is the interest rate?"))
    years = float(input("How many years do you plan on living in the house?"))
    if rate > .05 or years < 5: #If interest rate is greater than 5% or you will live there less than 5 years, output "You should rent"
        rent_or_buy = "You should rent"
    else: 
        rent_or_buy = "You should buy" #If interest rate is less than 5% or you will live there more than 5years, output "You should buy"
        
    money_after_housepayment = money_after_carloan - home_cost #Calculate remaining balance after monthly rent/mortgage payment
    return rent_or_buy, money_after_housepayment, home_cost #Return "Rent or Buy" statement, remaining balance, and monthly house payment
    
```

```python
rent_or_buy, money_after_housepayment, home_cost = housing(money_after_carloan)
print(f"It is recommeded: {rent_or_buy}")
print(f"The monthly house payment will be: ${home_cost}")
print(f"Balance after Roth IRA, Student Loan payment, Car Loan payment, and Monthly House payment: ${money_after_housepayment:.2f}")
```

8\. (4 pts) Americans spend on average 12.8\% of their budget on food. They also spend 55.7\% of their food budget on food away from home (FAFH) like dining out or takeout or food delivery.  https://www.escoffier.edu/blog/world-food-drink/consumer-dining-trend-statistics/

In this question create a function that calculates how many FAFH meals you could buy if your food spending habits are exactly like the average American above, use the `pay` variable from Question 2 as an argument and `balance` as your 2nd argument of the function.

Inside the function, use input to get the `avg_cost` of a meal when the user eats out ($20 for example). You will calculate how much you would spend total on FAFH based on the percentages from the start of this question multiplied by `pay`. Then divide `pay` by `avg_cost` to get the `num_meals`. Finally, calculate the money you have left after your TOTAL food budget (that 12.8\% value).

Return the `num_meals` and `money_left`.

```python
def food(balance, money_after_housepayment): #Food budget function
    
    avg_cost = float(input("What is the average cost of a meal when dining out?")) #Average cost of meal dining out
    total_food_budget = balance * .128 #12.8% of budget goes to food
    fafh_food_budget = total_food_budget * .557 #55.7% of food budget goes to dining out
    num_meals = int(fafh_food_budget / avg_cost) #How many dining out meals can you have based on your food budget and average cost to dine out
    money_after_food = money_after_housepayment - total_food_budget #Calculate balance after food budget is deducted
    
    return num_meals, money_after_food, total_food_budget, fafh_food_budget #Return number of dining out meals, total food budget, and dining out food budget
                 
```

```python
num_meals, money_left, total_food_budget, fafh_food_budget = food(balance, money_after_housepayment)
print(f"The monthly total food budget is: ${total_food_budget:.2f}")
print(f"The monthly fafh food budget is: ${fafh_food_budget:.2f}")
print(f"The number of monthly fafh meals is: {num_meals}")
print(f"Balance after Roth IRA, Student Loan payment, Car Loan payment, and Monthly House payment, and Food payment: ${money_left:.2f}")
```
