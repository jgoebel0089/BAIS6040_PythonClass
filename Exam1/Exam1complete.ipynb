```python
#Use the split function to break that string down into a list called limits.
limits = s.split()
print(limits)
```

```python
#Use list comprehension to convert each element in limits into a float.  [float(x) for…
limits = [float(y) for y in limits]
print(limits)
```

```python
#Take the max of limits – min of limits divided by the len of limits. This is the average increase to contribution limits each year. 
#Recently, they have increased by $500. Divide 500 by the average increase just found and round that product to an integer called interval.

average_limits_increase = (max(limits) - min(limits)) / len(limits) #Average limits increase per year recording limits
interval = int(round(500 / average_limits_increase)) #How many average increases per $500 increase

print(f"Average limits increase:, ${average_limits_increase:.2f}")
print("Interval:", interval)
```

```python
"""
Create a user-defined function named roth_simulator. The first argument is your percent_return, which is a percentage in the form of a decimal (0.1 for 10%). 
The 2nd argument is the interval when your max contribution increases by $500. Along with this, you’ll also have 2 default arguments: contribution = 7000 and years = 35.
"""

def roth_simulator(percent_return, interval, contribution = 7000, years = 35):
    catch_up = 0 #Will change to 1000 when years > 24, Assume user is 25 years old so after 50 you can contribute an additional $1000 to the balance
    balance = 0
    current_contribution = float(contribution)
    for i in range(years):
        #Increase base contribution by $500 when (i + 2) is divisible by interval
        #Hint – use % sign for remainder and the contribution has already been at $7000 for two years. 
        #I don't understand how this logic is applying to roth contributions, why is having contributions at $7000 for two years applicable?
        if interval > 0 and (i + 2) % interval == 0: #Second year account is open will be i=0, so (i+2)/3 leaves a remainder of 0
            current_contribution += 500.00 #Every time years+interval % interval == 0, increase current_contribution by $500.00
        
        #If your index variable > 24, then set the catch_up = 1000 (you’re then over 50 years old)
        if i > 24:
            catch_up = 1000.00
        
        #Add contribution to balance, Add catch_up to balance
        balance += current_contribution
        balance += catch_up

        #Increase your balance by your percent_return (Assume you invest on Jan 1st) and follow the formula: balance = balance * (1 + percent_return)
        balance *= (1.0 + percent_return)

    #Return balance from the function
    return balance
        

    
```

```python
#Call the roth_simulator function. The 2 inputs need to be the average return from the S&P 500 (sp) and the integer interval from Question 1. 
#Assign the bal variable as the balance returned.

bal = roth_simulator(sp, interval, contribution = 7000, years = 35)
```
